```python
import mat73
import numpy as np


filepath = r"C:\Users\Msfin\Christine_matlab_files_aug2024\SSTindivsomata_GLM.mat"
data_dict = mat73.loadmat(filepath)

print(f"data_dict['animal']['ShiftLrate'][0].shape: {data_dict['animal']['ShiftLrate'][0].shape}")
print(f"data_dict['animal']['ShiftR'][0].shape: {data_dict['animal']['ShiftR'][0].shape}")
print(f"data_dict['animal']['ShiftRunning'][0].shape: {data_dict['animal']['ShiftRunning'][0].shape}")
print(f"data_dict['animal']['ShiftV'][0].shape: {data_dict['animal']['ShiftV'][0].shape}")

num_animals = len(data_dict['animal']['ShiftLrate'])

data_dictionary = {}

for animal_idx in range(num_animals):
    neuron_list = []
    
    
    num_neurons = data_dict['animal']['ShiftR'][animal_idx].shape[2]
    
    for neuron_idx in range(num_neurons):
        neuron_data = []
        neuron_data.append(data_dict['animal']['ShiftLrate'][animal_idx]) 
        neuron_data.append(data_dict['animal']['ShiftR'][animal_idx][:, :, neuron_idx])  
        neuron_data.append(data_dict['animal']['ShiftRunning'][animal_idx])  
        neuron_data.append(data_dict['animal']['ShiftV'][animal_idx])  
        
        combined_matrix = np.stack(neuron_data, axis=1) 
        
        neuron_list.append(combined_matrix)

    data_dictionary[f'animal_{animal_idx + 1}'] = neuron_list

print(f"Shape of the combined matrix for the first neuron of animal_1: {data_dictionary['animal_1'][0].shape}")

print(f"Total number of animals in dictionary: {len(data_dictionary)}")
print(f"Number of neurons for animal_1: {len(data_dictionary['animal_1'])}")
```
